# Plotting Training Scores of Models

### Comments
Can see that the model, when trained on corrected masks, reaches a lower loss and higher dice score.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in scores
baseline_scores = np.load('../train_stats/baseline_longer.npy', allow_pickle=True).item()
updated_masks_scores = np.load('../train_stats/updated_masks_longer2.npy', allow_pickle=True).item()
```

```python
# Plot losses
plt.plot(baseline_scores['train_losses'], label='baseline train loss')
plt.plot(baseline_scores['val_losses'], label='baseline valid loss')

plt.plot(updated_masks_scores['train_losses'], label='updated masks train loss')
plt.plot(updated_masks_scores['val_losses'], label='updated masks valid loss')

plt.legend()

plt.xlabel('Epochs')
plt.ylabel('BCE + Dice Loss')

plt.show()
```

```python
# Plot dice scores
plt.plot(baseline_scores['train_dice_scores'], label='baseline train dice')
plt.plot(baseline_scores['val_dice_scores'], label='baseline valid dice')

plt.plot(updated_masks_scores['train_dice_scores'], label='updated masks train dice')
plt.plot(updated_masks_scores['val_dice_scores'], label='updated masks valid dice')

plt.legend()

plt.xlabel('Epochs')
plt.ylabel('Dice Score')

plt.show()
```

```python
# Dice score at lowest loss?
updated_masks_scores['val_dice_scores'][np.argmin(updated_masks_scores['val_losses'])]
```
